```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('HE_Data2.xlsx')
df
```

```python
df['Geography']
```

```python
df.isnull().sum()
```

```python
teachingtype = df[['Teaching Institution type', 'Total']]
teachinginstitution = teachingtype.groupby('Teaching Institution type').sum()
```

```python
teachinginstitution
```

```python
import seaborn as sns
```

```python
teachinginstitution.values
```

```python
np.reshape(teachinginstitution.values,(1,-1))[0]
```

```python
p=sns.color_palette("hls",3)
```

```python
sns.barplot(x=teachinginstitution.index,y=np.reshape(teachinginstitution.values,(1,-1))[0],palette=p)
```

```python
studentno = df[['Area', 'Total']]
perarea = studentno.groupby('Area').sum()
perarea
```

```python
perarea.values
```

```python
np.reshape(perarea.values,(1,-1))[0]
```

```python
plt.figure(figsize=(25,20))
p=sns.color_palette("hls",len(perarea))
sns.barplot(y=perarea.index,x=np.reshape(perarea.values,(1,-1))[0], orient='h',palette=p)
```

```python
five_institution = df[df['Institution Name'].isin(['Aston University', 'Bath College', 'Bolton College', 'University of East London', 'University of Sussex'])]
five_institution
```

```python
five_institution.columns
```

```python
num_entrant = five_institution[['Institution Name', 'Agriculture and related subjects', 'Architecture Building and Planning', 'Biological Sciences', 'Business and Administrative studies', 'Computer Science', 'Creative Arts and Design', 'Education', 'Engineering and Technology', 'Geographical Studies', 'Historical and Philosophical studies', 'Initial Teacher Training', 'Languages', 'Law', 'Mass Communications and Documentation', 'Mathematical Sciences', 'Medicine and Dentistry', 'Social Studies', 'Physical Sciences', 'Subjects allied to Medicine', 'Veterinary Sciences', 'Combined']]
entrant = num_entrant.groupby('Institution Name').sum()
entrant
```

```python
et = entrant.transpose()
et
```

```python
et.plot(kind='bar', stacked=True, figsize=(20, 10))
plt.show()
```

```python
entrant.plot(kind='bar', stacked=True, figsize=(20, 10))
plt.show()
```

```python
average_value = df[['Teaching Institution type', 'Agriculture and related subjects', 'Architecture Building and Planning', 'Biological Sciences', 'Business and Administrative studies', 'Computer Science', 'Creative Arts and Design', 'Education', 'Engineering and Technology', 'Geographical Studies', 'Historical and Philosophical studies', 'Initial Teacher Training', 'Languages', 'Law', 'Mass Communications and Documentation', 'Mathematical Sciences', 'Medicine and Dentistry', 'Social Studies', 'Physical Sciences', 'Subjects allied to Medicine', 'Veterinary Sciences', 'Combined']]
average = average_value.groupby('Teaching Institution type').mean()
average
```

```python
av = average.transpose()
av
```

```python
average.index
```

```python
average.plot(kind='bar', stacked=True, figsize=(20, 10))
plt.show()
```

```python
av.plot(kind='bar', stacked=True, figsize=(20, 10))
plt.show()
```
